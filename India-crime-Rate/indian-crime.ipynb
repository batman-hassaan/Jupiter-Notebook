```python
for col in df.columns:
    if col != 'STATE/UT':
        df[col] = pd.to_numeric(df[col], errors='coerce')  # converts to float, handles errors as NaN
```

```python
plt.figure(figsize=(16, 12))  # Adjust figure size as needed

sns.heatmap(df.corr(numeric_only=True),  cmap='coolwarm', fmt=".2f" , cbar=True)

plt.title("Correlation Heatmap of Crime Statistics")
plt.tight_layout()
plt.show()
```

```python
top_states = df.groupby('STATE/UT')['TOTAL IPC CRIMES'].mean().sort_values(ascending=False).head(10).index
df_filtered = df[df['STATE/UT'].isin(top_states)]

# Create line plot
fig = px.line(
    df_filtered,
    x='YEAR',
    y='TOTAL IPC CRIMES(per_100k)',
    color='STATE/UT',                # separate line for each state
    markers=True,
    title='YEAR vs. TOTAL IPC CRIMES(per_100k) across Top States',
    labels={
        'YEAR': 'Year',
        'TOTAL IPC CRIMES(per_100k)': 'Total IPC Crimes (per 100k)',
        'STATE/UT': 'State/UT'
    }
)

# Optional: customize
fig.update_traces(mode='lines+markers')
fig.update_layout(hovermode='x unified')  # hover shows all states at same year

fig.show()
```

```python
top_states = df.groupby('STATE/UT')['TOTAL IPC CRIMES'].mean().sort_values(ascending=False).head(10).index

# Filter data to only those states
df_filtered = df[df['STATE/UT'].isin(top_states)]

# Encode STATE/UT as numeric values for y-axis, keep mapping for labels
state_to_num = {state: idx for idx, state in enumerate(sorted(df_filtered['STATE/UT'].unique()))}
df_filtered['STATE_NUM'] = df_filtered['STATE/UT'].map(state_to_num)

# Build 3D scatter plot
fig = go.Figure()

fig.add_trace(
    go.Scatter3d(
        x=df_filtered['YEAR'],
        y=df_filtered['STATE_NUM'],
        z=df_filtered['TOTAL IPC CRIMES(per_100k)'],
        mode='markers+lines',
        text=df_filtered['STATE/UT'],   # hover will show actual state
        marker=dict(size=4, color=df_filtered['TOTAL IPC CRIMES(per_100k)'], colorscale='Viridis'),
    )
)

# Update layout: show state names instead of numbers
fig.update_layout(
    title='YEAR vs. STATE/UT vs. TOTAL IPC CRIMES(per_100k)',
    scene=dict(
        xaxis=dict(title='Year'),
        yaxis=dict(
            title='STATE/UT',
            tickmode='array',
            tickvals=list(state_to_num.values()),
            ticktext=list(state_to_num.keys())
        ),
        zaxis=dict(title='TOTAL IPC CRIMES(per_100k)')
    ),
    margin=dict(l=0, r=0, b=0, t=30)
)

fig.show()
```

```python
heatmap_data = df.pivot_table(
    index="STATE/UT", 
    columns="YEAR", 
    values="RAPE(per_100k)",  # Replace with "MURDER(per_100k)" or other crime
    aggfunc="mean"  # Use mean to handle duplicates (if any)
)

# Plot heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(
    heatmap_data, 
    cmap="YlOrRd",  # Yellow -> Red color scale
    annot=True,      # Show values in cells
    fmt=".1f",       # Format decimals
    linewidths=0.5,
    annot_kws={"size": 8}
)

# Customize
plt.title("State vs. Year Heatmap of Rape Rate (per 100k)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("State/UT")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate by state (average over years or pick last year)
df_latest = df[df['YEAR'] == df['YEAR'].max()]  # use latest year
# Or average over years:
# df_latest = df.groupby('STATE/UT')[['MURDER(per_100k)', 'ATTEMPT TO MURDER(per_100k)']].mean().reset_index()

# Melt data for grouped bar
df_melted = df_latest.melt(
    id_vars='STATE/UT',
    value_vars=['MURDER(per_100k)', 'ATTEMPT TO MURDER(per_100k)'],
    var_name='Crime Type',
    value_name='Rate'
)

# Plot
fig = px.bar(
    df_melted,
    x='STATE/UT',
    y='Rate',
    color='Crime Type',
    barmode='group',
    title='MURDER vs. ATTEMPT TO MURDER (per 100k) by State (latest year)',
    labels={'Rate': 'Rate (per 100k)', 'STATE/UT': 'State/UT'}
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Aggregate by state: average over years (to avoid year noise)
df_grouped = df.groupby('STATE/UT')['IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES(per_100k)'].mean().reset_index()

# Sort descending & pick top N (optional)
df_grouped = df_grouped.sort_values(by='IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES(per_100k)', ascending=False).head(10)

# Plot
fig = px.bar(
    df_grouped,
    x='STATE/UT',
    y='IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES(per_100k)',
    title='Average Rate of Importation of Girls from Foreign Countries (per 100k)',
    labels={'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES(per_100k)': 'Importation Rate (per 100k)', 'STATE/UT': 'State/UT'},
    color='IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES(per_100k)',
    text='IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES(per_100k)'  # show number on bar
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
